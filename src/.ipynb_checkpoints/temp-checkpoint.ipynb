```python
# calculate sums & average of goals of each team
df = match[["home_team_id", "away_team_id", "home_team_goal", "away_team_goal"]]
home_team_goal = pd.DataFrame(df.groupby(by = "home_team_id").sum()["home_team_goal"])
home_team_goal["home_count"] = df.groupby(by = "home_team_id").count()["away_team_id"]
away_team_goal = pd.DataFrame(df.groupby(by = "away_team_id").sum()["away_team_goal"])
away_team_goal["away_count"] = df.groupby(by = "away_team_id").count()["home_team_id"]

team_goal = home_team_goal.join(away_team_goal, how = 'outer')
team_goal["total_goal"]= team_goal.home_team_goal + team_goal.away_team_goal
team_goal["total_count"] = team_goal.home_count + team_goal.away_count
team_goal["average_goal"] = team_goal["total_goal"]/team_goal["total_count"]
team_goal = team_goal[["total_goal","total_count", "average_goal"]]
team_goal.index.names = ["team_id"]
```

```python
# calculate mean number of wins of each team
df = match1[["home_team_id", "away_team_id", "outcome"]]
df["away_outcome"]= -df["outcome"]

home_team_outcome = pd.DataFrame(df.groupby(by = "home_team_id").sum()["outcome"])
home_team_outcome["num_matches_home"] = df.groupby(by = "home_team_id").count()["away_team_id"]
away_team_outcome = pd.DataFrame(df.groupby(by = "away_team_id").sum()["away_outcome"])
away_team_outcome["num_matches_away"] = df.groupby(by = "away_team_id").count()["home_team_id"]
team_outcome = home_team_outcome.join(away_team_outcome, how = 'outer')

team_outcome["total_win"]= team_outcome.outcome + team_outcome.away_outcome
team_outcome["total_count"] = team_outcome.num_matches_home + team_outcome.num_matches_away
team_outcome["average_win"] = team_outcome["total_win"]/team_outcome["total_count"]
team_outcome = team_outcome[["total_win","total_count", "average_win"]]
team_outcome.index.names = ["team_id"]
```
